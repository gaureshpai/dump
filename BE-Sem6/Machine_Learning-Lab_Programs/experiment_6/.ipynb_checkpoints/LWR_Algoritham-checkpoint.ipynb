```python
# Use 'total_bill' to predict 'tip'
X = data["total_bill"].values.reshape(-1, 1)
y = data["tip"].values
```

```python
# Standardize X
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Define simple Locally Weighted Regression (LWR)
def lwr(X_train, y_train, X_test, tau):
    y_pred = np.zeros(len(X_test))
    for i, x in enumerate(X_test):
        weights = np.exp(-np.sum((X_train - x) ** 2, axis=1) / (2 * tau ** 2))
        W = np.diag(weights)
        X_b = np.c_[np.ones_like(X_train), X_train]  # Add bias term
        theta = np.linalg.pinv(X_b.T @ W @ X_b) @ (X_b.T @ W @ y_train)
        y_pred[i] = np.r_[1, x] @ theta  # Prediction
    return y_pred
```

```python
# Predict using LWR
tau = 0.5  # Bandwidth parameter
X_test = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
```

```python
y_pred = lwr(X, y, X_test, tau)
y_pred
```

```python
# Plot results
plt.scatter(X, y, color="blue", label="Data points", alpha=0.6)
plt.plot(X_test, y_pred, color="red", label="LWR Fit", linewidth=2)
plt.xlabel("Total Bill (Standardized)")
plt.ylabel("Tip")
plt.title("Locally Weighted Regression on Tips Dataset")
plt.legend()
plt.show()
```

```python
# ---- Predict for new data ----
new_bills = np.array([[10], [20], [30], [40]])  # New total bill values
new_bills_scaled = scaler.transform(new_bills)  # Standardize new data
predicted_tips = lwr(X, y, new_bills_scaled, tau)
print("Predicted Tips:", predicted_tips)
```
